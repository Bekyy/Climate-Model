```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.projections import PolarAxes
import mpl_toolkits.axisartist.grid_finder as gf
import mpl_toolkits.axisartist.floating_axes as fa

class TaylorDiagram(object):
    def __init__(self, STD, rect=111, label='_'):
        self.STD = STD
        tr = PolarAxes.PolarTransform()
        rlocs = np.concatenate(((np.arange(11.0) / 10.0), [0.95, 0.99]))
        tlocs = np.arccos(rlocs)
        gl1 = gf.FixedLocator(tlocs)
        tf1 = gf.DictFormatter(dict(zip(tlocs, map(str, rlocs))))
        self.smin = 0
        self.smax = 8.5 * self.STD
        gh = fa.GridHelperCurveLinear(tr, extremes=(0, (np.pi / 2), self.smin, self.smax), grid_locator1=gl1,
                                      tick_formatter1=tf1)
        fig = plt.figure()
        ax = fa.FloatingSubplot(fig, rect, grid_helper=gh)
        fig.add_subplot(ax)
        ax.axis['top'].set_axis_direction('bottom')
        ax.axis['top'].label.set_text("Correlation coefficient")
        ax.axis['top'].label.set_size(17)  # Set the font size to 12
        ax.axis['top'].toggle(ticklabels=True, label=True)
        ax.axis['top'].major_ticklabels.set_axis_direction('top')
        ax.axis['top'].label.set_axis_direction('top')
        ax.axis['left'].set_axis_direction('bottom')
        ax.axis['left'].label.set_text(" ")
        ax.axis['left'].toggle(ticklabels=True, label=True)
        ax.axis['left'].major_ticklabels.set_axis_direction('bottom')
        ax.axis['left'].label.set_axis_direction('bottom')
        ax.axis['right'].set_axis_direction('top')
        ax.axis['right'].label.set_text("Standard deviation")
        ax.axis['right'].label.set_size(17)  # Set the font size to 12
        ax.axis['right'].toggle(ticklabels=True, label=True)
        ax.axis['right'].major_ticklabels.set_axis_direction('left')
        ax.axis['right'].label.set_axis_direction('top')
        ax.axis['bottom'].set_visible(False)
        ax.grid()
        self._ax = ax
        self.ax = ax.get_aux_axes(tr)
        l, = self.ax.plot([0], self.STD, 'k*', ls='', ms=12, label=label)
        t = np.linspace(0, (np.pi / 2.0))
        r = np.zeros_like(t) + self.STD
        self.ax.plot(t, r, 'k--', label='_')
        self.samplePoints = [l]

    def add_sample(self, STD, r, label, marker, color, *args, **kwargs):
        l, = self.ax.plot(np.arccos(r), STD, marker, color=color, *args, label=label, **kwargs)
        self.samplePoints.append(l)
        return l

    def add_contours(self, levels=5, **kwargs):
        rs, ts = np.meshgrid(np.linspace(self.smin, self.smax), np.linspace(0, (np.pi / 2.0)))
        RMSE = np.sqrt(np.power(self.STD, 2) + np.power(rs, 2) - (2.0 * self.STD * rs * np.cos(ts)))
        contours = self.ax.contour(ts, rs, RMSE, levels, **kwargs)
        return contours

    def show(self):
        plt.show()

def srl(obsSTD, df, colors, fname):
    fig = plt.figure(figsize=(8, 8))
    dia = TaylorDiagram(obsSTD, label='ref')
    plt.title('Minimum Temprature',fontsize=20)
    plt.clabel(dia.add_contours(colors='#808080'), inline=1, fontsize=10)

    for index, row in df.iterrows():
        dia.add_sample(row['STD'], row['Correlation'], label=f"{row['Label']} (RMSE: {row['STD']:.2f})",
                       marker='^', color=row['Color'], mec=row['Color'], mfc=row['Color'], mew=1.6)

    # Adding legend for climate models outside the plot area
    climate_model_legend = [
        plt.Line2D([0], [0], marker='o', color='w', label='CCCMA', markerfacecolor=colors[0], markersize=10, markeredgecolor=colors[0]),
        plt.Line2D([0], [0], marker='o', color='w', label='CNRM', markerfacecolor=colors[1], markersize=10, markeredgecolor=colors[1]),
        plt.Line2D([0], [0], marker='o', color='w', label='MIROC', markerfacecolor=colors[2], markersize=10, markeredgecolor=colors[2]),
        plt.Line2D([0], [0], marker='o', color='w', label='MPI', markerfacecolor=colors[3], markersize=10, markeredgecolor=colors[3]),
        plt.Line2D([0], [0], marker='o', color='w', label='EM', markerfacecolor=colors[4], markersize=10, markeredgecolor=colors[4]),
    ]
    plt.legend(handles=climate_model_legend,title='RCP Models',title_fontsize=15, fontsize=12,loc='upper left', bbox_to_anchor=(1, 1))

    plt.savefig(fname, bbox_inches='tight')
    plt.show()

# Example DataFrame for the new data
data = {
    'Label': ['CCCMA', 'CNRM', 'MIROC', 'MPI', 'EM'],
    'STD': [1.977311, 1.984637, 2.125935, 1.815345,1.676762],
    'Correlation': [0.648657,0.583629, 0.506580, 0.595694,0.685731],
    'Color': ['blue', 'green', 'orange', 'brown', 'red']
}

df = pd.DataFrame(data)
df

# Calling the function with the new data
obsSTD = 1
colors = ['blue', 'green', 'orange', 'brown', 'red']  # Add a new color for MPI
fname = 'TaylorDiagram.jpg'
srl(obsSTD, df, colors, fname)
```
